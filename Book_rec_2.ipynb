# Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books, on='book_id')
```

```python
num_rating_df = ratings_with_name.groupby('title').count()['rating'].reset_index()
num_rating_df.rename(columns={'rating': 'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('title')['rating'].mean().reset_index()
avg_rating_df.rename(columns={'rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='title')
```

```python
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books, on='title').drop_duplicates('title')[['title', 'authors', 'small_image_url', 'num_ratings', 'avg_rating']]
```

```python
popular_df
```

# Collaborative Filtering Recommending System

```python
x = ratings_with_name.groupby('user_id').count()['rating'] > 100
```

```python
learned_users = x[x].index
learned_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['user_id'].isin(learned_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('title').count()['rating'] >=100
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='title', columns='user_id', values='rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1], reverse=True) [1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```
