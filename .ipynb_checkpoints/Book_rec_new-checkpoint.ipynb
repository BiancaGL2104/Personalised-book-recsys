# Model Based Collaborative Filtering

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torch import nn
```

```python
merged_df['user_id'] = merged_df['user_id'].astype("category")
merged_df['title'] = merged_df['title'].astype("category")
```

```python
df_surprise1 = merged_df[['user_id', 'title', 'rating']].copy()
df_surprise1.columns = ['user_id', 'title', 'rating']
```

```python
df_surprise1['reviews'] = df_surprise1.groupby('title')['rating'].transform('count')
```

```python
df_surprise1['reviews'] = df_surprise1.groupby('title', observed=True)['rating'].transform('count')
```

```python
df_surprise1 = df_surprise1[df_surprise1['reviews'] > 100]
```

```python
class RatingDataset(Dataset):
    def __init__(self, df, user_col, item_col, rating_col):
        self.users = torch.tensor(df[user_col].cat.codes.values, dtype=torch.long)  # Ensure tensor is long
        self.items = torch.tensor(df[item_col].cat.codes.values, dtype=torch.long)  # Ensure tensor is long
        self.ratings = torch.tensor(df[rating_col].values.astype(np.float32))
    
    def __len__(self):
        return len(self.ratings)
    
    def __getitem__(self, idx):
        return self.users[idx], self.items[idx], self.ratings[idx]
```

```python
dataset = RatingDataset(df_surprise1, 'user_id', 'title', 'rating')
dataloader = DataLoader(dataset, batch_size=512, shuffle=True)
```

```python
class MatrixFactorization(nn.Module):
    def __init__(self, num_users, num_items, embedding_size=20):
        super(MatrixFactorization, self).__init__()
        self.user_embeddings = nn.Embedding(num_users, embedding_size)
        self.item_embeddings = nn.Embedding(num_items, embedding_size)
        # Initialize embeddings
        self.user_embeddings.weight.data.uniform_(0, 0.05)
        self.item_embeddings.weight.data.uniform_(0, 0.05)
        
    def forward(self, user_ids, item_ids):
        user_ids = user_ids.long()  # Ensure IDs are long
        item_ids = item_ids.long()  # Ensure IDs are long
        user_embedding = self.user_embeddings(user_ids)
        item_embedding = self.item_embeddings(item_ids)
        return (user_embedding * item_embedding).sum(1)
```

```python
num_users = df_surprise1['user_id'].cat.categories.size
num_items = df_surprise1['title'].cat.categories.size
model = MatrixFactorization(num_users, num_items)

# Define loss function and optimizer
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
# Training loop
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    for users, items, ratings in dataloader:
        predictions = model(users, items)
        loss = loss_fn(predictions, ratings)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    print(f"Epoch {epoch}: Loss = {loss.item()}")
```

```python
# Assume user ID for recommendation
target_user_id = 1001

# Predict ratings for all items by this user
all_items = torch.tensor(np.arange(num_items), dtype=torch.long)
all_users = torch.full((num_items,), target_user_id, dtype=torch.long)

# Switch model to evaluation mode
model.eval()
with torch.no_grad():
    predicted_ratings = model(all_users, all_items)

# Get the items that the user has already rated
known_items = df_surprise1[df_surprise1['user_id'] == target_user_id]['title'].unique()

# Filter out known items
recommended_items = [(item, rating) for item, rating in enumerate(predicted_ratings) if item not in known_items]

# Sort the predictions and get the top 10 recommendations
recommended_items.sort(key=lambda x: x[1], reverse=True)
top_ten_recommended_items = recommended_items[:10]

print("Top 10 recommended items for user 1001:")
print(top_ten_recommended_items)
```

```python
books = books[['ISBN','title']]
books.head()
ratings.rename(columns={'book_rating':'rating'},inplace=True)
ratings.head()
min_book_ratings = 50 #a book was rated at least 
min_user_ratings =  50 #a user rated books at least


ratings_flrd_df = ratings.groupby("ISBN").filter(lambda x: x['ISBN'].count() >= min_book_ratings)
ratings_flrd_df = ratings_flrd_df.groupby("userID").filter(lambda x: x['userID'].count() >= min_user_ratings)



"{0} movies deleted; all movies are now rated at least: {1} times. Old dimensions: {2}; New dimensions: {3}"\
.format(len(ratings.ISBN.value_counts()) - len(ratings_flrd_df.ISBN.value_counts())\
        ,min_book_ratings,ratings.shape, ratings_flrd_df.shape )
ratings_flrd_df['rating'].value_counts()
ratings_flrd_df.head()
reader = Reader(rating_scale=(1,10)) #line_format by default order of the fields
data = Dataset.load_from_df(ratings_flrd_df[["userID",	"ISBN",	"rating"]], reader=reader)

trainset = data.build_full_trainset()

testset = trainset.build_anti_testset()
algo_SVD = SVD(n_factors = 11)
algo_SVD.fit(trainset)


# Predict ratings for all pairs (i,j) that are NOT in the training set.
testset = trainset.build_anti_testset()

predictions = algo_SVD.test(testset)

# subset of the list  predictions
predictions[0:2]
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from collections import defaultdict #data colector

#Surprise: https://surprise.readthedocs.io/en/stable/
import surprise

from surprise.reader import Reader
from surprise import Dataset
from surprise.model_selection import GridSearchCV

  ##CrossValidation
from surprise.model_selection import cross_validate


  ##Matrix Factorization Algorithms
from surprise import SVD
from surprise import NMF

np.random.seed(42) # replicating results
def get_top_n(predictions, userID, books, ratings, n = 10):
    '''Return the top N (default) ISBN for a user,.i.e. userID and history for comparison
    Args:
    Returns: 
  
    '''
    #Part I.: Surprise docomuntation
    
    #1. First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    #2. Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key = lambda x: x[1], reverse = True)
        top_n[uid] = user_ratings[: n ]
    
    #Part II.: inspired by: https://beckernick.github.io/matrix-factorization-recommender/
    
    #3. Tells how many books the user has already rated
    user_data = ratings[ratings.userID == (userID)]
    print('User {0} has already rated {1} books.'.format(userID, user_data.shape[0]))

    
    #4. Data Frame with predictions. 
    preds_df = pd.DataFrame([(id, pair[0],pair[1]) for id, row in top_n.items() for pair in row],
                        columns=["userID" ,"ISBN","rat_pred"])
    
    
    #5. Return pred_usr, i.e. top N recommended books with (merged) titles. 
    pred_usr = preds_df[preds_df["userID"] == (userID)].merge(books, how = 'left', left_on = 'ISBN', right_on = 'ISBN')
            
    #6. Return hist_usr, i.e. top N historically rated books with (merged) titles for holistic evaluation
    hist_usr = ratings[ratings.userID == (userID) ].sort_values("rating", ascending = False).merge\
    (books, how = 'left', left_on = 'ISBN', right_on = 'ISBN')
    
    
    return hist_usr, pred_usr
hist_SVD_277427, pred_SVD_277427 = get_top_n(predictions, books = books, userID = 277427, ratings = ratings)
hist_SVD_277427.head(15)
pred_SVD_277427
```

```python
# Pivot table to create a user-item matrix
user_item_matrix = ratings_flrd_df.pivot_table(index='user_id', columns='book_id', values='rating', fill_value=0)

# Function to recommend books
def recommend_books(userID, num_recommendations=10):
    if userID not in user_item_matrix.index:
        print(f"User ID {userID} not found in the data.")
        return None

    # Extract user ratings and identify unrated books
    user_ratings = user_item_matrix.loc[userID]
    unrated_books = user_ratings[user_ratings == 0]

    if unrated_books.empty:
        print(f"No unrated books available for user ID {userID}.")
        return None

    # Sort unrated books based on some external criterion (here, random for demonstration)
    recommended_books = unrated_books.sample(min(num_recommendations, len(unrated_books)))

    # Merge with the books DataFrame to get book titles
    recommended_books_df = recommended_books.index.to_frame(name='book_id').reset_index(drop=True)
    recommendations = recommended_books_df.merge(books_df, on='book_id', how='left')
    return recommendations[['book_id', 'title']]

# Example use
recommended_books_for_user = recommend_books(1002, 10)
if recommended_books_for_user is not None:
    print(recommended_books_for_user)
```
